```python
def sgd(x, y, alpha = 0.1, itmax = 10):
    
    n = len(y)
    w0 = 0
    w1 = 0
    yhat = np.zeros(n)
    e = np.zeros(n)
    indices = np.arange(n)
    for t in range(itmax):
        np.random.shuffle(indices)
        for j in indices:
            yhat[j] = (w1 * x[j]) + w0
            e[j] = y[j] - yhat[j]
            w0 = w0 + (alpha * e[j])
            w1 = w1 + (alpha * e[j]) * x[j]
        print("t = ", t)
        print("yhat = ", yhat)
        print("e = ", e)
        print("w = ", w0, ", ", w1, "\n")
    
    
    MSE = ((e ** 2).sum())/(2*n)
    print(w0, w1, MSE)
    return w0, w1, MSE

# Generate data
x = np.array([1, 2, 3, 4])
y = np.array([3, 5, 7, 9])

# Train model and get parameters
w0, w1, mse = sgd(x, y, itmax=10)

# Create plot
fig, ax = plt.subplots()

# Plot data points
ax.scatter(x, y, label='Data')

# Plot predicted line
x_range = np.linspace(x.min(), x.max(), 100)
y_pred = w1 * x_range + w0
ax.plot(x_range, y_pred, label='Model')

# Add axis labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('SGD Linear Regression')

# Show plot
plt.show()
```

```python
def ols(x, y):
    n = len(y)
    x = np.c_[np.ones(n), x]
    w = np.linalg.solve((x.T @ x),(x.T @ y)).reshape(-1,1)
    yhat = x @ w
    e = y - yhat
    MSE = ((e ** 2).sum())/(2*n)
    return w, MSE

# Generate data
x = np.array([1, 2, 3, 4])
y = np.array([3, 5, 7, 9])

# Train model and get parameters
w, mse = ols(x, y)

# Create plot
fig, ax = plt.subplots()

# Plot data points
ax.scatter(x, y, label='Data')

# Plot predicted line
x_range = np.linspace(x.min(), x.max(), 100)
y_pred = w[1] * x_range + w[0]
ax.plot(x_range, y_pred, label='Model')

# Add axis labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('OLS Linear Regression')

# Show plot
plt.show()
```
